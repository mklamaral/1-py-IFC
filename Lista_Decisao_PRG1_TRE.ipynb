# 1. Ler um valor e escrever se é positivo, negativo ou zero.

```python
# 1. Ler um valor e escrever se é positivo, negativo ou zero.

# Entrada
a = float(input("Digite um número: "))

# Processamento
if a>0:
  txt = ("\npositivo")
elif a==0:
  txt =("\nzero")
else:
  txt =("\nnegativo")

# Saida
print(txt)
```

# 2. Ler um valor e escrever a mensagem É MAIOR QUE 10! se o valor lido for maior que 10, caso contrário escrever NÃO É MAIOR QUE 10!

```python
# 2. Ler um valor e escrever a mensagem É MAIOR QUE 10! se o valor lido for maior que 10, caso contrário escrever NÃO É MAIOR QUE 10!

# entrada
a = float(input("Digite um número: "))

# processamento
if a>10:
  txt = "É MAIOR QUE 10!"
elif a==10:
  txt = "É 10!"
else:
  txt = "NÃO É MAIOR QUE 10!"

# saida
print(txt)

```

# 3. Calcule a soma de dois números, se o resultado for maior que 10, mostre-o na tela.

```python
# 3. Calcule a soma de dois números, se o resultado for maior que 10, mostre-o na tela.

# entrada

a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))

# processamento
c = a+b
if c>=10:
  print(f"A soma dos valores é: {c}")

```

# 4. Entrar com um número e informar se ele é divisível por 5.

```python
# 4. Entrar com um número e informar se ele é divisível por 5.

# entrada
a = float(input("Digite um número: "))

# processamento
b = (a%5)

if b == 0:
  txt = f"O numero {a:.2f} é divisível por 5!"
else:
  txt = f"O numero {a:.2f} não é divisível por 5!"

# saída
print(txt)
```
